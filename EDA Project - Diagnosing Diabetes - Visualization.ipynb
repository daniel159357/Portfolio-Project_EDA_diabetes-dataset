```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#2 preview using .head
diabetes_data = pd.read_csv("diabetes.csv")
print(diabetes_data.head())
```

```python
# Pairplot by Matplotlob 
pairplot = pd.plotting.scatter_matrix(diabetes_data, figsize=(20,20), marker = 'o', hist_kwds = {'bins': 15}, s = 10, alpha = 0.8)
plt.show()

# https://stackoverflow.com/questions/56188305/matplotlib-to-plot-a-pairplot
# https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.scatter.html?highlight=scatter
```

```python
# Pairplot by Seaborn
plt.close()
sns.pairplot(diabetes_data, hue="Outcome")
plt.show()
```

```python
# Heatmap by Seaborn
plt.close()
sns.heatmap(diabetes_data.corr(), center=0,robust=True, annot=True, fmt=".1f", cmap="Blues")
plt.show()

# https://matplotlib.org/stable/users/explain/colors/colormaps.html
# https://seaborn.pydata.org/tutorial/color_palettes.html
```

There is no set value for what counts as “highly correlated”, however a general rule is a correlation of 0.7 (or -0.7). There are no pairs of features with a correlation of 0.7 or higher, so we do not need to consider leaving any features out of our model based on multicollinearity.
